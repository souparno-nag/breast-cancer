```python
df.drop(columns=['ICCC3WHO', 'ICCC3XWHO', 'IHS'], inplace=True)
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df['is_alive'] = (df['STAT_REC'] == 1).astype(int)
df.drop(columns=['STAT_REC','CODKM','ICD_5DIG'], inplace=True)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('is_alive', axis=1))
df_scaled = pd.DataFrame(scaled_features, columns=df.drop('is_alive', axis=1).columns)
df_scaled['is_alive'] = df['is_alive'].values
df = df_scaled
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
X = df.drop('is_alive', axis=1)
y = df['is_alive']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
# Correlation

def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE, SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt

print("="*60)
print("FINDING COMMON FEATURES FROM RFE AND ANOVA")
print("="*60)
```

```python
print("\n1. Applying ANOVA F-test...")
anova_selector = SelectKBest(score_func=f_classif, k=20)
X_train_anova = anova_selector.fit_transform(X_train, y_train)

anova_features = X_train.columns[anova_selector.get_support()]
print(f"ANOVA selected features ({len(anova_features)}): {list(anova_features)}")
```

```python
from sklearn.linear_model import SGDClassifier

print("\n2. Applying RFE (with SGDClassifier)...")

# Lightweight linear classifier
model = SGDClassifier(loss='log_loss', max_iter=500, random_state=42)

# RFE with fewer features and bigger step for speed
rfe_selector = RFE(estimator=model, n_features_to_select=20, step=2)
X_train_rfe = rfe_selector.fit_transform(X_train, y_train)

rfe_features = X_train.columns[rfe_selector.support_]
print(f"RFE selected features ({len(rfe_features)}): {list(rfe_features)}")
```

```python
print("\n3. Finding common features...")
common_features = list(set(anova_features) & set(rfe_features))
print(f"Common features ({len(common_features)}): {common_features}")
```

```python
if len(common_features) > 0:
    X_train_common = X_train[common_features]
    X_test_common = X_test[common_features]

    print(f"\nDataset shapes after selecting common features:")
    print(f"X_train_common: {X_train_common.shape}")
    print(f"X_test_common: {X_test_common.shape}")
```

```python
anova_only = list(set(anova_features) - set(rfe_features))
rfe_only = list(set(rfe_features) - set(anova_features))

print(f"\nFeatures selected by ANOVA only ({len(anova_only)}): {anova_only}")
print(f"Features selected by RFE only ({len(rfe_only)}): {rfe_only}")
```

```python
models = {
    'ANOVA_only': LogisticRegression(max_iter=1000, random_state=42),
    'RFE_only': LogisticRegression(max_iter=1000, random_state=42),
    'Common_features': LogisticRegression(max_iter=1000, random_state=42),
    'All_features': LogisticRegression(max_iter=1000, random_state=42)
}

datasets = {
    'ANOVA_only': (X_train[anova_features], X_test[anova_features]),
    'RFE_only': (X_train[rfe_features], X_test[rfe_features]),
    'Common_features': (X_train_common, X_test_common),
    'All_features': (X_train, X_test)
}

results = {}

for name, (X_tr, X_te) in datasets.items():
    print(f"\nTraining {name} model...")
    print(f"Features used: {X_tr.shape[1]}")
    
    cv_scores = cross_val_score(models[name], X_tr, y_train, cv=5, scoring='accuracy')
    models[name].fit(X_tr, y_train)
    y_pred = models[name].predict(X_te)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    results[name] = {
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'test_accuracy': test_accuracy,
        'n_features': X_tr.shape[1]
    }
    
    print(f"CV Accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
    print(f"Test Accuracy: {test_accuracy:.4f}")
```

# Ensemble 4

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

# Bagging with shallower trees and fewer estimators
bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3),  # shallower tree
    n_estimators=5,                                 # fewer trees
    random_state=42
)

# Gradient Boosting with fewer estimators and lower learning rate
boosting = GradientBoostingClassifier(
    n_estimators=50,       # fewer estimators
    learning_rate=0.05,    # more stable training
    max_depth=2,           # shallower boosting trees
    random_state=42
)

# Inner stack with lighter models
stack_inner = StackingClassifier(
    estimators = [
        ('linear-svc', LinearSVC(C=1.0, max_iter=10000, random_state=42)),
        ('knn', KNeighborsClassifier(
            n_neighbors=2,
            weights='distance',
            metric='euclidean'
        ))
    ],
    final_estimator=LogisticRegression(max_iter=100)
)

# Final ensemble stacking the reduced models
final_ensemble = StackingClassifier(
    estimators=[
        ('bag', bagging),
        ('boost', boosting),
        ('stack', stack_inner)
    ],
    final_estimator=LogisticRegression(max_iter=100)
)

# Fit on training data
final_ensemble.fit(X_train_common, y_train)

# Evaluate
from sklearn.metrics import accuracy_score
y_pred = final_ensemble.predict(X_test_common)
print(f"Final Ensemble Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```
